<font size="4"><b>1.1. Introduction</b></font> <a class="anchor" id="section_1.1"></a>

---

To reduce the final energy demand from the transport sector, two complementary strategies are considered: sufficiency and efficiency.
These strategies are structured into six levers, which guide the transition toward a low-energy mobility system.

Sufficiency refers to a reduction in the need for energy-intensive transport activity. It is achieved through behavioral and systemic changes that avoid or reduce car use, including modal shifts and shared mobility.

Efficiency refers to technical improvements that reduce the amount of energy consumed per kilometre travelled or per vehicle, without necessarily reducing mobility demand.

#### a) Travelled Distance (Gpkm/year) 

The travelled distance represents the total number of kilometres covered annually by a person using all land-based transport modes. It includes commuting, leisure, and other mobility needs.

In 2019, this value amounts to **151,9 Gpkm/year** [3], which corresponds to an average of **13200 km/person/year**. 

$\textcolor{Orange}{\textbf{Assumption:}}$ In line with typical negaWatt france friends (2017), a **15% reduction** is expected by 2050, driven by the expansion of teleworking and a higher share of urban housing. This leads to an estimated **130,18 Gpkm/year** , equivalent to **10915 km/person/year** [3] in 2050. 


#### b) Modal Shift

The travelled distance is the total number of kilometres covered by a person in a year using all land-based transport modes, including soft mobility options such as walking and cycling.

In 2019, **74,75%** of all passenger-kilometres are made by car [3], while the remaining 25.25% are shared among public transport, cycling, walking, and other modes.

$\textcolor{Orange}{\textbf{Assumption:}}$ The objective is to **reduce car modal share to 48,3%** by 2050 [3] through increased use of soft and collective mobility solutions.
> **Comment**: ici c'est pas vraiment une assumption comme c'est une conséquence des assumptions des modal share (voir section suivante), donc à revoir comment le noté dans le résumé de negaWatt... (mais pour moi c'est faut)


#### c) Car Occupancy

Car occupancy refers to the average number of people travelling in a car simultaneously. It directly impacts the number of vehicle-kilometres required to meet passenger demand.

In 2019, the average occupancy rate is **1.25 persons per car** [5].

$\textcolor{Orange}{\textbf{Assumption :}}$ With the development of carpooling and shared mobility services, this figure is expected to increase to **rise it to 2 persons per car** by 2050 [5].


#### d) Speed Limit & Driving Behaviour 

In 2019, no specific policies were implemented to reduce speeds or encourage eco-driving. The reference energy efficiency gain is therefore set to **0%** in the base year. 

However, smoother driving and lower speed limits reduce aerodynamic drag and rolling resistance, contributing to reduced energy consumption per kilometre.

$\textcolor{Orange}{\textbf{Assumption:}}$ From 2030, new speed regulations will be introduced across all regions: 
- 110 km/h on highways (currently 120 km/h),
- 70 km/h on regional roads,
- 30 km/h in urban areas.
According to the négaWatt France scenario, reducing highway speed from 130 km/h to 110 km/h, combined with more sustainable driving behaviour, leads to a 3% reduction in fuel consumption. In Belgium, since the initial limit is already 120 km/h, the expected reduction in energy use is estimated at **2%** [1].

This change is assumed to occur in a single step in 2030, and its impact is not explicitly modelled, but rather approximated within the global vehicle efficiency improvement described in the next section.

> **Comment**: This reduction seems really small. It should be checked again for a better reference!

#### e) Vehicle Design (Weight Reduction)
The evolution of vehicle design, especially the reduction in average weight, contributes to lower energy consumption per kilometre.

$\textcolor{Orange}{\textbf{Assumption:}}$ From 2035, a progressive decrease in vehicle mass will lead to a **2% reduction** in energy consumption by 2050 [1].

Like speed policies, this effect is not explicitly isolated in the model but is approximated as part of the global efficiency improvement of electric vehicles.

> **Comment**: The assumption of a 2% energy reduction by 2050 due to vehicle weight is taken from Mobility analysis.xlsx [1], but does not appear in any other confirmed source. It is loosely attributed to négaWatt France, but lacks clear justification. Further validation is recommended.

<font size="5"><b>2. Modal shares</b></font> <a class="anchor" id="section_2"></a>

---

This section defines the modal shares used to allocate total land-based passenger-kilometers across transport modes. These shares are expressed in percentages and must sum to 100% for each year. Air transport is excluded, while walking and cycling (soft modes) are explicitly included in the distribution.

> **Comment**: For most modes, the projections were made starting from 2015 JRC data and extended to 2050. Now that updated JRC statistics are available up to 2021, it’s clear that the 2019 estimates used as a baseline for cars are inaccurate. To stay aligned with the rest of the analysis (which begins in 2019), the 2015 values are not displayed here, but this also highlights the need to revise the entire car breakdown using the most recent data.

The percentage values for 2015 are based on JRC [3] and the soft mode are based on Monitor [7]. The objective for the following years is to progressively reduce the proportion of cars in favour of more sustainable modes of transport. The expected evolution of each mode is defined by the following assumptions.

#### a) Soft modes [7]:
- **Pedestrians**: It is assumed that the proportion of people walking will increase slightly, from 2.00% to 3.00%, by 2050. This reflects ongoing urban policies that promote pedestrian zones and walkable city centres.
> **Comment**: This assumption may be conservative, given that trends in pedestrianisation could lead to a more significant shift in dense urban areas.
- **Bicycle**: The proportion of people cycling increased from 4.70% to 12.00%, which is a growth of 2.5 times. This growth is supported by the rapid adoption of electric bikes and major investments in cycling infrastructure, including dedicated cycle lanes and secure bike parking.

#### b) Rail [3]:
- **Tram/metro**: It is assumed that there will be a slight increase from 0.95% to 1.20%.
> **Comment**: This may be a conservative estimate. Given the development of new tram lines in Liège and other cities, a higher increase could be expected, particularly if coordinated with upgrades to other modes of public transport.
- **Train**: The train share is projected to double, from 7.30% to 14.00%. This aligns with national strategies such as SNCB's Vision Rail 2040 Belgium [6] (actualization May 2022). SNCB projects that trains could reach a modal share of 15% by 2040, excluding soft modes. When walking and cycling are included, this equates to around 13%, which lends weight to the plausibility of our target of 14% by 2050. 

#### c) Road [3]:
- **Motorcycle**: Its share rises from 1.30% to 3.50%, which is more than double. This could be due to an increased use of two-wheelers for commuting in urban and peri-urban areas, particularly given the growing availability of low-emission or electric models.
- **Bus/coach**: The share has doubled from 9.00% to 18.00%. This growth is considered plausible if rural and suburban areas are served by an expanded network of high-frequency bus services.

- **Car**: As a direct consequence of the increases in all other modes, the car modal share drops from 74.75% in 2019 to 48.30% in 2050, representing a 35% reduction.

```python
# Define target years and initial/final modal shares
years = np.array([2019, 2025, 2030, 2035, 2040, 2045, 2050])
modes = {
    'Pedestrian': (2.00, 3.00),
    'Bicycle': (4.70, 12.00),
    'Motorcycle': (1.30, 3.50),
    'Bus/coach': (9.00, 18.00),
    'Tram/metro': (0.95, 1.20),
    'Train': (7.30, 14.00),
}

# Prepare DataFrame
df_modal = pd.DataFrame(index=list(modes.keys()), columns=years, dtype=float)
t = np.linspace(0, 1, len(years))

# Generate modal share curves
for mode, (start, end) in modes.items():
    if mode == 'Train':
        values = []
        for year in years:
            if year <= 2040:
                val = start + (13.0 - start) * (year - 2019)/(2040 - 2019)
            else:
                val = 13.0 + (end - 13.0) * (year - 2040)/(2050 - 2040)
            values.append(round(val, 2))
        df_modal.loc[mode] = values                                      # curve adjusted to hit 13% at 2040
    elif mode in ['Pedestrian', 'Bicycle', 'Motorcycle']:          
        df_modal.loc[mode] = np.linspace(start, end, len(years))  # Linear growth
    else :                                                         
        df_modal.loc[mode] = start + (end - start) * (3 - 2*t) * t**2    # Accelerated growth

df_modal.loc['Car'] = (100 - df_modal.sum()).round(2)   # Calculate 'Car' share as complement to 100%
df_modal = df_modal.round(2)  # Round all values for readability                           
df_modal
```


---

<font size="4"><b>2.1. Final dataframe </b></font> <a class="anchor" id="section_2_1"></a>

---

```python
# 1) Compute Gpkm and km per person from modal shares and sufficiency data
df_gpkm = pd.DataFrame({year: df_suff["Passenger_km [Gpkm/year]"][year] * df_modal[year] / 100 for year in years}, index=df_modal.index).round(2)

df_kmp = pd.DataFrame({year: df_gpkm[year] * 1e9 / population_dict[year] for year in years}, index=df_modal.index).round(2)

# 2) Build MultiIndex for Mode and Unit
arrays = [np.repeat(df_modal.index, 3), ['% du total', 'Gpkm', 'km/per'] * len(df_modal)]
mi = pd.MultiIndex.from_arrays(arrays, names=['Mode', 'Unit'])

# 3) Interleave percentage, Gpkm, and km/per for each mode
data_rows = []
for mode in df_modal.index:
    data_rows.append(df_modal.loc[mode].values)  # modal percentages
    data_rows.append(df_gpkm.loc[mode].values)   # Gpkm values
    data_rows.append(df_kmp.loc[mode].values)    # km per person
data = np.vstack(data_rows)

# 4) Create DataFrame and reset index for styling
final = pd.DataFrame(data, index=mi, columns=years)
fr = final.reset_index()

# 5) Append TOTAL row with fixed values
total_gpkm = df_suff["Passenger_km [Gpkm/year]"]
total_kmp = df_suff["Km per person [km/person/year]"]
rows = []
for unit in ['% du total', 'Gpkm', 'km/per']:
    if unit == '% du total':
        vals = [100] * len(years)
    elif unit == 'Gpkm':
        vals = [total_gpkm[year] for year in years]
    else:
        vals = [total_kmp[year] for year in years]
    rows.append(pd.DataFrame([['TOTAL', unit, *vals]], columns=fr.columns))
fr = pd.concat([fr] + rows, ignore_index=True)
mode_full = fr['Mode'].tolist()
    
# 6) Only show mode name on the first line of each block
fr['Mode'] = np.where(fr['Unit'] == '% du total', fr['Mode'], '')

# 7) Define styling functions for key highlights and borders
def highlight_key(df):
    mask = pd.DataFrame('', index=df.index, columns=df.columns)
    for i, unit in enumerate(fr['Unit']):
        real_mode = mode_full[i] 
        if real_mode=='TOTAL' and unit == '% du total':
            mask.iloc[i, 2:] = 'background-color:#ffe5b4;font-weight:bold'
        if (real_mode=='Car' and unit=='% du total') or (real_mode=='TOTAL' and unit=='Gpkm'):
            mask.iloc[i, df.columns.get_indexer([2019,2050])] = 'background-color:#ffcccc;font-weight:bold'
    return mask

def highlight_lines(row):
    if row['Unit'] == '% du total':
        return ['border-top:2px solid black'] * len(row)
    if row['Mode'] == 'TOTAL':
        return ['border-top:4px double black'] * len(row)
    return [''] * len(row)

# 8) Apply styles and display
styled = (
    fr.style
      .apply(highlight_key, axis=None)
      .apply(highlight_lines, axis=1)
      .set_properties(subset=['Mode'], **{'font-weight':'bold'})
      .set_properties(subset=['Unit'], **{'font-style':'italic','color':'gray'})
      .format({year:"{:.2f}" for year in years})
      .set_caption("Modal Shares")
      .hide(axis='index')
      .set_table_attributes('style="width:100%;table-layout:fixed;"')
)

display(styled)
```

<font size="4"><b>3.2. Car </b></font> <a class="anchor" id="section_3_2"></a>

---

This mode includes all types of car, with shares expressed as a percentage of the total Gpkm for cars.

$\textcolor{Orange}{\textbf{Assumption:}}$

- **BEV cars**: 0.5% in 2020 → 99.5% in 2050, based on the electrification trajectory defined in section 1.3
- All other types decrease proportionally, starting from their 2019 shares
- The 2019 values were estimated using JRC 2015 data and projected backwards to 2050.

The following method is used to compute all car types except BEVs:
$$
\text{Share}_{y} = \text{Share}_{2019} \times \frac{100 - \text{BEV}_{y}}{99.5}
$$

```python
# Car
# Initial 2019 shares 
initial_shares = {
    "Gasoline cars":         35.3,
    "Diesel cars":           62.5,
    "LPG cars":               0.2,
    "NGV / biomethane cars":  0.3,
    "Hydrogen cars":          0.0,
    "Plug-in gasoline":       1.1,
    "Plug-in diesel":         0.1,
    "Plug-in LPG":            0.0,
}

car_shares = {}
for mode, init in initial_shares.items():
    projected = []
    for year in years:
        bev_pct = df_techno.loc[year, "Electrification [%]"]
        projected.append(init * (100 - bev_pct) / 99.5)
    car_shares[mode] = projected
car_shares["BEV cars"] = df_techno["Electrification [%]"].tolist()
df_cars = pd.DataFrame(car_shares, index=years).T.round(3)
```

<font size="4"><b>4.1. Modal shares</b></font> <a class="anchor" id="section_4_1"></a>

---

The modal shares are based on JRC data from 2015 [3], but have not been updated to reflect 2019 values. The 2015 values are not shown as the rest of the study begins in 2019. The aim is to reduce road freight by increasing rail and inland waterway transport. 

#### a) Rail
In 2019, rail accounted for 10% of the total. Based on 'Vision Rail 2040' [6], the expected target is 20%. However, this scenario intentionally remains below the 20% target in order to preserve a significant role for inland waterways. Here, the share increases to 19% in 2040 and remains constant until 2050.
> **Comment**: This constant evolution over ten years does not seem realistic to me, as real systems evolve more progressively.


#### b) Inland waterways
The proportion of waterborne freight is expected to grow from 15% in 2019 to 23% in 2050. Assuming a gradual and continuous increase, this growth would counterbalance the reduction in road freight without placing excessive strain on the rail network.
> **Comment**: The statistics from the JRC show that the trend is decreasing. In addition, this value does not take into account domestic coastal shipping.

#### c) Road freight
Road freight decreases from 75% in 2020 to 58% in 2050. The shift away from road is motivated by the need to decarbonize the sector and reduce energy consumption. It is disaggregated in two ways:
- by **type of route**:
    - **National** freight decreases from 35% to 27% between 2020 and 2050.
    - **International** freight drops from 40% to 31% over the same period.
- by **vehicle type**:
    - **LCVs** (light commercial vehicles) remain stable at around 1.5–1.6%, reflecting their continued use for short-distance and last-mile logistics.
    - **HDVs** (heavy-duty vehicles), which dominate the road freight segment, see a reduction from 73.5% in 2020 to 56.4% in 2050, consistent with the overall decrease in road freight.

Both breakdowns refer to the same total share of road freight and cannot be added together. The objective is to reduce long-distance freight, especially that carried by HDVs, which can more easily be shifted to rail or water. LCVs remain dominant for short urban deliveries and are considered to be a constant factor.

#### d) Total freight 
The total freight activity remains stable around 75 Gtkm from 2020 to 2035, before dropping to 70 Gtkm from 2040 onwards.
> **Comment**: this sudden 5 Gtkm drop in just five years seems abrupt and not clearly justified by the scenario, it may require adjustment.

```python
# Modal share (%)
rail = pd.Series([10.0, 12.0, 14.0, 17.0, 19.0, 19.0, 19.0], index=years)
water = pd.Series([15.0, 16.0, 18.0, 20.0, 22.0, 23.0, 23.0], index=years)
road_total = 100 - (rail + water)

national = pd.Series([35.0, 34.0, 32.0, 29.0, 26.5, 27.0, 27.0], index=years)
international = road_total - national
hdv = pd.Series([73.5, 70.5, 66.5, 61.5, 57.4, 56.4, 56.4], index=years)
lcv = road_total - hdv

# Total freight volumes in Gtkm, convert to string
total_freight = pd.Series([75.0, 75.0, 75.0, 75.0, 70.0, 70.0, 70.0], index=years).map(lambda x: f"{x:.2f}")

# Build DataFrames
df_rail_freight = pd.DataFrame({'Rail (%)': rail.values}, index=years).T
df_water = pd.DataFrame({'Water (%)': water.values}, index=years).T
df_road_total = pd.DataFrame({'Road (%)': road_total.values}, index=years).T
df_road_route = pd.DataFrame({'National (%)': national.values, 'International (%)': international.values}, index=years).T
df_road_vehicle = pd.DataFrame({'HDV (%)': hdv.values, 'LCV (%)': lcv.values}, index=years).T
# Create freight DataFrame with strings
df_total_freight = pd.DataFrame({'Total freight (Gtkm)': total_freight.values}, index=years).T


# Combine in desired order
mode_dfs = {
    'Rail': df_rail_freight,
    'Water': df_water,
    'Road': df_road_total,
    'Road (by route)': df_road_route,
    'Road (by vehicle)': df_road_vehicle,
    'Total freight (Gtkm)': df_total_freight,
}

rows = []
for mode, df in mode_dfs.items():
    temp = df.copy()
    temp['Mode'] = mode
    temp['Type'] = temp.index
    temp = temp.reset_index(drop=True)
    rows.append(temp)

combined = pd.concat(rows, ignore_index=True)
combined['Mode'] = combined['Mode'].mask(combined['Mode'].duplicated(), '')
fr = combined[['Mode', 'Type'] + list(years)]

# Styling
def highlight_mode_separator(row):
    return ['border-top:2px solid black'] * len(row) if row['Mode'] != '' else [''] * len(row)

def bold_mode(cell, mode_cell, col):
    return 'font-weight:bold' if mode_cell and col == 'Mode' else ''

styled = fr.style.hide(axis='index')\
    .set_caption("Carrier Shares and Freight Volume")\
    .set_table_attributes('style="width:100%;table-layout:fixed;"')\
    .apply(highlight_mode_separator, axis=1)\
    .apply(lambda row: [bold_mode(cell, row['Mode'], col) for col, cell in zip(fr.columns, row)], axis=1)\
    .set_properties(subset=['Type'], **{'font-style':'italic'})

# Apply percentage formatting only to rows ending with '(%)'
mask = fr['Type'].str.endswith('(%)')
for year in years:
    styled = styled.format({year: "{:.1f}%"}, subset=(mask, year))

styled
```

<font size="5"><b>6. Conversion to TWh </b></font> <a class="anchor" id="section_6"></a>

---

In this section, I focus on generating the input data required by the PyPSA-Eur model for the transport sector. More specifically, I extract and format the values used in the script "Clever_data_code.py" (source: [10]).

These data correspond to final energy consumption (FEC) per final energy carrier and are used as exogenous inputs in the PyPSA-Eur model. They are derived directly from the energy values used in the scenario assessment in the previous section ('modal share' and 'carrier share'), but are expressed in different units (e.g. TWh, Gtkm) and are based on several assumptions.


---

<font size="4"><b>6.1. Passenger road mobility </b></font> <a class="anchor" id="section_6_1"></a>

---

$\textcolor{steelblue}{\textbf{Simplification:}}$

At this stage, only data from cars is considered. Other modes of passenger transport, such as buses and motorcycles, are excluded.

The aim is to express final energy consumption per carrier in TWh/year using previously calculated demand in passenger kilometres (pkm), assuming an average energy intensity per car technology.

> **comment**: Buses and motorcycles are included in energy demand calculations. However, buses will represent ~20% of total passenger transport by 2050 yet aren't used in current PyPSA inputs. To me, this is not negligible.

```python
# --- Energy consumption [kWh/km] for all car types ---

# Energy content per unit  --> Sources??
energy_content = {
    "Gasoline": 9.21,          # [kWh/L]
    "Diesel": 10.17,           # [kWh/L]
    "LPG": 6.69,               # [kWh/L]
    "NGV / biomethane": 11.3,  # [kWh/kg]
    "Hydrogen": 33.3           # [kWh/kg]
}

# Base consumption in fuel unit per 100 km  --> Sources??
consumption_per_100km = {
    "Gasoline": 7.151,         # [L/100km]
    "Diesel": 5.947,           # [L/100km]
    "LPG": 9.402,              # [L/100km]
    "NGV / biomethane": 4.182, # [kg/100km]
    "Hydrogen": 1.127          # [kg/100km]
}

# Electrification share for PHEVs
share_electric = 0.70

# Convert to kWh/km
df_consumption_kWh_km = pd.DataFrame({
    fuel: (consumption_per_100km[fuel] * energy_content[fuel]) / 100
    for fuel in consumption_per_100km
}, index=years).T

# BEV efficiency (data from "df_techno" provided in the section 1.2 efficiency)
df_bev = df_techno["Efficiency [kWh/km]"].to_frame().T
df_bev.index = ["BEV car"]

# Compute hybrid consumption
df_phev_consumption = pd.DataFrame({
    f"Plug-in_{fuel}": df_consumption_kWh_km.loc[fuel] * (1 - share_electric) + df_bev.loc["BEV car"] * share_electric
    for fuel in ["Gasoline", "Diesel", "LPG"]
}, index=years).T

# Combine
df_combined = pd.concat([df_consumption_kWh_km, df_phev_consumption, df_bev])
df_combined.columns.name = "Year"



# --- Travel demand [Gpkm] for all car types ---

# Get Gpkm for cars (data from "df_gpkm" provided in the section 1.1 sufficiency)
gpkm_car = df_gpkm.loc["Car"]

# Multiply car shares by Gpkm total for cars (data from "df_cars" provided in the section 3.2 car)
gpkm_by_car_type = df_cars*gpkm_car/100



# --- Final Energy Demand By Car Type (TWh) ---

# Compute energy consumption demand in TWh (data from "df_suff" provided in the section 1.1 sufficiency)
energy_demand_TWh = gpkm_by_car_type.values * df_combined.values / df_suff["Car occupancy [pers/car]"].values

# Rebuild the final DataFrame
energy_demand_TWh = pd.DataFrame(energy_demand_TWh, index=gpkm_by_car_type.index, columns=years).round(2)
energy_demand_TWh.loc["Total road passenger"] = energy_demand_TWh.sum(axis=0)

# put the impact of the speed limit & weight for each year (data from "df_suff" provided in the section 1.1 sufficiency) 
energy_demand_TWh = energy_demand_TWh* (1 + df_suff["Speed limit gain [%]"] / 100) * (1 + df_suff["Weight reduction gain [%]"] / 100)

def highlight_pypsa_inputs(row):
    if row.name in ["BEV cars", "Total road passenger", "Rail freight", "Final Energy Consumption [TWh]"]:
        return ['background-color: #ffcccc; font-weight: bold'] * len(row)
    return [''] * len(row)

styled = (
    energy_demand_TWh
    .style
    .apply(highlight_pypsa_inputs, axis=1)
    .format("{:.2f}")
    .set_caption("Final Energy Demand for Passenger Cars [TWh/year]")
)

display(styled)
```

$\textcolor{steelblue}{\textbf{Simplification:}}$

The current PyPSA inputs only use two rows from the 'Final Energy Demand for Passenger Cars ' table.

- "Total" (final energy consumption).
- "BEV car" (electricity consumption for cars).

Data is sampled every 10 years instead of every 5 years to reduce complexity.
The other car types will be considered later, but with a different assumption (see Section 7).

> **Comment**: A final correction is applied to energy_demand_TWh for speed limits and weight reduction. However, these adjustments appear to be included in the efficiency assumptions in Section 1.
 -> This suggests potential double-counting and requires clarification.

```python
# Energy efficiency values [kWh/100 pkm] --> source ??
efficiency_train = [9.927, 9.610, 9.310, 9.070, 8.871, 8.696, 8.516]

# Compute total Gpkm (data from "df_gpkm" provided in the section 1.1 sufficiency)
gpkm_total = df_gpkm.loc["Train"] + df_gpkm.loc["Tram/metro"]

# Convert to TWh
# Final energy = Gpkm * (kWh / 100 pkm) = TWh
final_energy_TWh = gpkm_total * efficiency_train / 100 

# Multiply total FEC by electric share (data from "df_train" provided in the section 3.5 Train)
electric_fec = final_energy_TWh * df_train.loc["Electric"] / 100

# Build final DataFrame
df_train_fec = pd.DataFrame([final_energy_TWh.values], index=["Total rail passenger"], columns=years).round(3)
df_train_fec.loc["Rail electricity"] = electric_fec.values.round(3)
df_train_fec
```

$\textcolor{steelblue}{\textbf{Simplification:}}$

These are the only two values that we use in Pypsa-EUR, but with a 10-year step.

> **Comment**: The value for Rail electricity is computed by applying the train electrification share to the total energy demand of rail (train + tram/metro).
However, this ignores the fact that trams and metros are fully electric, meaning the actual electricity share should be slightly higher (but negligeable).

<font size="4"><b>6.5. Air transport </b></font> <a class="anchor" id="section_6_5"></a>

---

$\textcolor{red}{\textbf{Attention:}}$

The final energy consumption for air travel used in the PyPSA inputs is based on Gpkm values that differ from those presented in Section 4 of the analysis (based on JRC data).

In particular:
- The value used for 2020 is 50 Gpkm, whereas the JRC source indicates a value closer to 37 Gpkm.
- This results in higher total energy values in the short term (e.g. 2030) and a more optimistic reduction by 2050 (down to ~16 Gpkm and ~4 TWh).
- Although domestic flights represent a minimal share (e.g. 0.038 Gpkm in 2030) and are set to zero by 2050, they are included in the calculation.

This discrepancy may not be critical, as the year 2020 is not actually used in the PyPSA model.

> **Comment**: Nevertheless, this discrepancy is undocumented, and the logic behind the updated Gpkm trajectory is unclear. This affects the trends and deletes the automation of this work. If the data source or assumptions are not justified, they may need to be updated for consistency.

```python
# Efficiency [kWh/100 pkm]
efficiency_air = pd.Series([37.300, 34.471, 32.817, 31.016, 29.068, 27.373, 25.916], index=years)

# PyPSA values (from dashboard V5.38)
gpkm_air = {
    'Domestic':    [0.053, 0.046, 0.038, 0.029, 0.020, 0.010, 0.000],
    'International': [48.195, 50.963, 45.454, 39.215, 32.181, 24.256, 15.353]
}
df_gpkm_air = pd.DataFrame(gpkm_air, index=years)
df_gpkm_air["Total"] = df_gpkm_air.sum(axis=1)

# Final energy consumption [TWh] 
fec_air = df_gpkm_air["Total"] * efficiency_air / 100

# Assemble comparison
df_air_table = pd.DataFrame({
    "Passenger-km (JRC-based projection)": df_air.loc["Total"],
    "Passenger-km (used in PyPSA)": df_gpkm_air["Total"],
    "Final Energy Consumption [TWh]": fec_air.round(3)
}).T

styled = (
    df_air_table
    .style
    .apply(highlight_pypsa_inputs, axis=1)
    .format("{:.2f}")
    .set_caption("Comparison between JRC-based and PyPSA air transport projections")
)

styled
```

Only final energy consumption is taken as input, and only for the years 2030, 2040 and 2050.
